```python
import pandas as pd
```

```python
fem_act = pd.read_csv('fem_act.csv')
fem_temp = pd.read_csv('fem_temp.csv')
male_act = pd.read_csv('male_act.csv')
male_temp = pd.read_csv('male_temp.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Constants
minutes_per_day = 1440
estrus_days = [2, 6, 10, 14]
estrus_minutes = []

# Convert estrus days to minute indices (0-indexed)
for day in estrus_days:
    start = (day - 1) * minutes_per_day
    end = day * minutes_per_day
    estrus_minutes.extend(range(start, end))

# Filter out estrus minutes from female data
fem_act_non_estrus = fem_act.drop(index=estrus_minutes)
fem_temp_non_estrus = fem_temp.drop(index=estrus_minutes)

# Compute mean activity and temp per individual
fem_act_means = fem_act_non_estrus.mean()
fem_temp_means = fem_temp_non_estrus.mean()
mal_act_means = male_act.mean()
mal_temp_means = male_temp.mean()

# Plotting boxplots
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Activity comparison
axs[0].boxplot([fem_act_means, mal_act_means], labels=["Females", "Males"])
axs[0].set_title("Average Activity (Non-Estrus Days)")
axs[0].set_ylabel("Activity Level")

# Temperature comparison
axs[1].boxplot([fem_temp_means, mal_temp_means], labels=["Females", "Males"])
axs[1].set_title("Average Temperature (Non-Estrus Days)")
axs[1].set_ylabel("Core Body Temperature")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Constants
minutes_per_day = 1440
estrus_days = [2, 6, 10, 14]
estrus_minutes = []

# Calculate estrus minute indices (0-based)
for day in estrus_days:
    start = (day - 1) * minutes_per_day
    end = day * minutes_per_day
    estrus_minutes.extend(range(start, end))

# Create masks
all_minutes = range(fem_temp.shape[0])
non_estrus_minutes = list(set(all_minutes) - set(estrus_minutes))

# Split female data
fem_estrus = fem_temp.loc[estrus_minutes]
fem_non_estrus = fem_temp.loc[non_estrus_minutes]

# Split male data (same indices for fairness)
mal_estrus = male_temp.loc[estrus_minutes]
mal_non_estrus = male_temp.loc[non_estrus_minutes]

# Mean temperature per mouse
fem_estrus_means = fem_estrus.mean()
fem_non_estrus_means = fem_non_estrus.mean()
mal_estrus_means = mal_estrus.mean()
mal_non_estrus_means = mal_non_estrus.mean()

# Plotting
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Estrus plot
bplot1 = axs[0].boxplot([fem_estrus_means, mal_estrus_means], labels=["Females", "Males"], patch_artist=True)
axs[0].set_title("Temperature During Estrus Days")
axs[0].set_ylabel("Core Body Temperature")

# Non-estrus plot
bplot2 = axs[1].boxplot([fem_non_estrus_means, mal_non_estrus_means], labels=["Females", "Males"], patch_artist=True)
axs[1].set_title("Temperature During Non-Estrus Days")
axs[1].set_ylabel("Core Body Temperature")

# Apply colors: pink for females, blue for males
colors = ["pink", "lightblue"]
for patch, color in zip(bplot1['boxes'], colors):
    patch.set_facecolor(color)
for patch, color in zip(bplot2['boxes'], colors):
    patch.set_facecolor(color)

plt.tight_layout()
plt.show()
```
